```python
import pandas as pd

df = pd.read_excel('higher_edu_total_2019_filter.xlsx')

#White Men 
WA_Man = df.loc[(df['State'] == Maryland) & (df['B2_Gender'] == 2) & (df['B4.1_Race'] == 6)]
WA_Women = df.loc[(df['B3.2_Country'] == 0) & (df['B2_Gender'] == 1) & (df['B4.1_Race'] == 6)]

# Kickstart the xlsxwriter
writer = pd.ExcelWriter('WhiteAmerican.xlsx', engine='xlsxwriter')
WA_Man.to_excel(writer, sheet_name='Sheet1', index=False)
WA_Women.to_excel(writer, sheet_name='Sheet2', index=False)

# Finally write the file
writer.save()
```
