```python
#Loading the library 
import pandas as pd
import plotly.express as px 
import altair as alt
import numpy
import numpy as np
```

```python
# Loading the dataset (average morgage rate is calculated every year)
df= pd.read_excel('Mortgage_data.xlsx') 
```

# Mortgage rates (Americas, 1971-2022)

```python
import plotly.express as px
fig = px.scatter(df, x='Year', y='Mortgage', color='Mortgage', trendline="lowess", render_mode ='auto', labels= {"Mortgage" : "Mortgage rates", 'Year': 'Year (1971-2022)'})

#If 'lowess’, a Locally Weighted Scatterplot Smoothing line will be drawn for each discrete-color/symbol group.

# Correct position of x points
for scatter, trendline in zip(fig.data[::2], fig.data[1::2]):
    trendline['x'] = scatter['x'][np.logical_not(np.isnan(scatter['y']))]

fig.update_traces(textposition='top center')
    
fig.update_layout(
    height=700,
    title_text='Mortgage rates (Americas, 1971-2022)'
)

fig.show()
```

```python
# plotting the scatter chart 
fig = px.line(df.dropna(), y='Mortgage', x = 'Year' , title = 'Mortgage rates for 30 years in the United States' ) 

# showing the plot 
fig.show()
```

```python
import plotly.express as px

fig = px.bar(df, x='Year', y='Mortgage',
             hover_data=['Mortgage', 'Year'], color='Mortgage',
             labels={'Mortgage'},text_auto=True, height=500)
fig.show()
```

```python
import plotly.express as px
fig = px.area(df.dropna(), x='Year', y='Mortgage',
             hover_data=['Mortgage', 'Year'],
             labels={'Mortgage'}, height=500)

fig.show()
```

```python
import plotly.express as px
fig = px.area(df.dropna(), x='Year', y='Mortgage',
             hover_data=['Mortgage', 'Year'],
             labels={'Mortgage'}, height=500, pattern_shape_sequence=["."])

fig.show()
```
